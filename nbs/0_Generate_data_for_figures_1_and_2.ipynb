```python
import numpy as np
import pandas as pd
```

# Figure 1 data

```python
n_years = 1000
```

```python
baseline_emission=np.full(n_years, 1)
baseline_emission[0]=0
```

```python
cdr_only = [0,-0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, 
            -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05]
co2_perturbation_pulse = np.r_[cdr_only, np.full(n_years, 0)]   # long tail of 0s
co2_cumulative_delay = np.convolve(cdr_only, np.ones(n_years, dtype=float), mode='full')
```

```python
accelerated_emission_only = np.arange(1.05,0,-0.05)
accelerated_emission_only[0]=np.nan
co2_accelerated_emission = np.r_[accelerated_emission_only, np.full(1000, 0)]   # long tail of 0s
```

```python
upfront_emissions_yr1=np.full(n_years, 3)
upfront_emissions_yr1[0]=0
upfront_removal_yr1=np.full(n_years, -1.15)
upfront_removal_yr1[0]=0
```

```python
embodied_trace=np.full(n_years, -1.15)
embodied_trace[22:]=0
embodied_trace[0]=0
embodied_trace[1]=3
```

```python
co2_cumulative_delay[0]=0
co2_accelerated_emission[0]=0
```

```python
df_emissions = pd.DataFrame(baseline_emission, columns=['baseline_emission'])
df_emissions['co2_cumulative_delay']=co2_cumulative_delay[:1000]
df_emissions['co2_accelerated_emission']=co2_accelerated_emission[:1000]
df_emissions['upfront_removal_yr1']=upfront_removal_yr1
df_emissions['upfront_emissions_yr1']=upfront_emissions_yr1
df_emissions.to_csv('emissions_time_series_pulse.csv')
```

# Figure 2 data

```python
baseline_20yr_pulse = np.convolve(baseline_emission, np.full(20, 1))[:1000]
df_emissions_20yr = pd.DataFrame(baseline_20yr_pulse, columns=['baseline_emission'])
df_emissions_20yr['co2_cumulative_delay']=np.convolve(co2_cumulative_delay, np.full(20, 1))[:1000]
df_emissions_20yr['net_accelerated_emission']=np.convolve(co2_cumulative_delay[:1000]+co2_accelerated_emission[:1000], np.full(20, 1))[:1000]
df_emissions_20yr['embodied']=np.cumsum(embodied_trace)[:1000]
df_emissions_20yr.to_csv('emissions_time_series_20yr.csv')
```
