```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from vis_settings import flux_types as colors
```

# Quick calculations

```python
n_years = 1000
```

```python
baseline_emission=np.full(n_years, 1)
baseline_emission[0]=0
```

```python
cdr_only = [0,-0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, 
            -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05, -0.05]
co2_perturbation_pulse = np.r_[cdr_only, np.full(n_years, 0)]   # long tail of 0s
co2_cumulative_delay = np.convolve(cdr_only, np.ones(n_years, dtype=float), mode='full')
```

```python
accelerated_emission_only = np.arange(1.05,0,-0.05)
accelerated_emission_only[0]=np.nan
co2_accelerated_emission = np.r_[accelerated_emission_only, np.full(1000, 0)]   # long tail of 0s
```

```python
upfront_emissions_yr1=np.full(n_years, 3)
upfront_emissions_yr1[0]=0
upfront_removal_yr1=np.full(n_years, -1.15)
upfront_removal_yr1[0]=0
```

# Figure 1

```python
lwidth=3
def format_subplot():
    plt.xlim([-3,100])
    plt.axhline(y=0,linestyle='--',color='k', alpha=0.3, linewidth=1)

plt.subplots(nrows=2,ncols=2, figsize=(12,8), sharex=True, sharey=True)

plt.subplot(2,2,1)
plt.plot(co2_cumulative_delay, linewidth=lwidth, color=colors['removal'])
plt.plot(baseline_emission, color='k',alpha=0.6, linewidth=lwidth)
plt.title("A) Physical delay")
format_subplot()

plt.subplot(2,2,2)
plt.plot(co2_cumulative_delay, linewidth=lwidth, color=colors['removal'])
plt.plot(baseline_emission, color='k',alpha=0.6, linewidth=lwidth)
plt.title("B) Counterfactual delay")
format_subplot()

plt.subplot(2,2,3)
plt.plot(baseline_emission, color='k',alpha=0.6, linewidth=lwidth)
plt.plot(co2_cumulative_delay, linewidth=lwidth, color=colors['removal'])
plt.plot(co2_accelerated_emission, linewidth=lwidth, color=colors['activity-emission'])
plt.title("C) Accelerated emissions")
format_subplot()

plt.subplot(2,2,4)
plt.plot(baseline_emission, color='k',alpha=0.6, linewidth=lwidth, label="Baseline Emissions")
plt.plot(upfront_removal_yr1, linewidth=lwidth, label="Removal", color=colors['removal'])
plt.plot(upfront_emissions_yr1, linewidth=lwidth, label="CDR Activity Emissions", color=colors['activity-emission'])
plt.legend()
plt.title("D) Upfront embodied emissions")
format_subplot()
```

# Figure 2

```python
embodied_trace=np.full(n_years, -1.15)
embodied_trace[22:]=0
embodied_trace[0]=0
embodied_trace[1]=3
np.nansum(embodied_trace)
```

```python
lwidth=3
def format_subplot():
    plt.xlim([-3,100])
    plt.axhline(y=0,linestyle='--',color='k', alpha=0.3, linewidth=1)

plt.subplots(nrows=2,ncols=2, figsize=(12,8), sharex=True, sharey=False)

plt.subplot(2,2,1)
plt.plot(np.convolve(co2_cumulative_delay, np.full(20, 1)), linewidth=lwidth, color=colors['net-effect'])
plt.plot(np.convolve(baseline_emission, np.full(20, 1)), color='k',alpha=0.6, linewidth=lwidth)
plt.title("A) Physical delay")
format_subplot()

plt.subplot(2,2,2)
plt.plot(np.convolve(co2_cumulative_delay, np.full(20, 1)), linewidth=lwidth, color=colors['net-effect'])
plt.plot(np.convolve(baseline_emission, np.full(20, 1)), color='k',alpha=0.6, linewidth=lwidth)
plt.title("B) Counterfactual delay")
format_subplot()

plt.subplot(2,2,3)
co2_cumulative_delay[0]=0
co2_accelerated_emission[0]=0
co2_cumulative_delay[:1000]+co2_accelerated_emission[:1000]
plt.plot(np.convolve(co2_cumulative_delay[:1000]+co2_accelerated_emission[:1000], np.full(20, 1)), linewidth=lwidth, color=colors['net-effect'])
plt.plot(np.convolve(baseline_emission, np.full(20, 1)), color='k',alpha=0.6, linewidth=lwidth)
plt.title("C) Accelerated emissions")
format_subplot()

plt.subplot(2,2,4)
plt.plot(np.cumsum(embodied_trace), linewidth=lwidth, label="Net CDR activity", color=colors['net-effect'])
plt.plot(np.convolve(baseline_emission, np.full(20, 1)), color='k',alpha=0.6, linewidth=lwidth, label="Baseline emission")
plt.legend()
plt.title("D) Upfront embodied emissions")
format_subplot()
```
